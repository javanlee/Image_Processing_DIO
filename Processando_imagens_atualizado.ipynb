```python
def img2gray(img):
  r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]
  gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
  return gray
```

```python
def imgpb(img):
  r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]
  pb = 0.5 * r + 0.5 * g + 0.5 * b
  minimo=np.min(pb)
  maximo=np.max(pb)
  for i in range(len(pb)):
    for j in range(len(pb[i])):
      if minimo <= pb[i,j] < (minimo +maximo/2.0):
        pb[i,j] = 0
      elif pb[i,j] >= (minimo +maximo/2.0):
        pb[i,j] = 255 # ou 1
      else:
        pb[i,j] = 0
  return pb
```

```python
image_test = plt.imread('cats_and_dogs/test/22.jpg')
gray_test = img2gray(image_test)
pb = imgpb(image_test)

f, axarr = plt.subplots(1,3)
axarr[0].imshow(image_test)
axarr[1].imshow(gray_test, cmap='gray')
axarr[2].imshow(pb, cmap='gray')


f.set_figwidth(10)

plt.show()
```
